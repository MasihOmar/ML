```python
# Load the dataset
df = pd.read_csv('/kaggle/input/turkish-poems/all_poems.csv')
```

```python
# Convert to lowercase and remove punctuation
df['poem_text'] = df['poem_text'].str.lower().str.replace('[^\w\s]','')
df['poem_text'] = df['poem_text'].apply(word_tokenize)
```

# **Search according to frequency**

```python
poet_name = input('Enter the name of the poet whose words and frequencies you want to count: ')
min_freq = int(input('Enter the minimum frequency of words to include: '))
poet_df = df.loc[df['poet_name'] == poet_name]
```

```python
if poet_df.empty:
    print('No poet found with name:', poet_name)
else:
    poet_text = str(poet_df['poem_text'].sum())
    poet_word_list = poet_text.split()
    word_counts = Counter(poet_word_list)
    filtered_word_counts = {word: count for word, count in word_counts.items() if count >= min_freq}
    print('Poet:', poet_name)
    print(f'Words used in their poems with frequency >= {min_freq}:')
    if len(filtered_word_counts) == 0:
        print('No words found with frequency >=', min_freq)
    else:
        for word, count in filtered_word_counts.items():
            print(word, ':', count)
```
